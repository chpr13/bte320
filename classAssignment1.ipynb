```python
sales = input("Enter sales ($): ")
taxRate = input("Enter tax rate (%): ")

sales = float(sales)
taxRate = float(taxRate)

print("Sales:     $: " , (f"{sales:,.2f}" ))

costOfGoods = sales * .40
print ("Cost of Goods: $", (f"{costOfGoods:,.2f}" ))

grossProfit = sales - costOfGoods
print ("Gross Profit: $", (f"{grossProfit:,.2f}" ))

taxes = grossProfit * (taxRate / 100)
print ("Taxes: $", (f"{taxes:,.2f}" ))

netIncome = grossProfit - taxes
print ("Net Income: $", (f"{netIncome:,.2f}" ))
```

```python
number1 = input("Enter first number: ")
number1 = float(number1)

number2 = input("Enter second number: ")
number2 = float(number2)

operator = input("Enter operator: ")


if operator == "+":
    print(number1 + number2)
elif operator == "-":
    print(number1 - number2)
elif operator == "*":
    print(number1 * number2)
elif operator == "/":
    print(number1 / number2)
elif operator == "%":
    print(number1 % number2)
elif operator == "**":
    print(number1 ** number2)
elif operator == "//":
    print(number1 // number2)

else:
    print("Invalid operator")
```
