```python
sales = input("Enter sales ($): ")
taxRate = input("Enter tax rate (%): ")

sales = float(sales)
taxRate = float(taxRate)

print("Sales:     $: " , (f"{sales:,.2f}" ))

costOfGoods = sales * .40
print ("Cost of Goods: $", (f"{costOfGoods:,.2f}" ))

grossProfit = sales - costOfGoods
print ("Gross Profit: $", (f"{grossProfit:,.2f}" ))

taxes = grossProfit * (taxRate / 100)
print ("Taxes: $", (f"{taxes:,.2f}" ))

netIncome = grossProfit - taxes
print ("Net Income: $", (f"{netIncome:,.2f}" ))
```

```python
number1 = input("Enter first number: ")
number1 = float(number1)

number2 = input("Enter second number: ")
number2 = float(number2)

operator = input("Enter operator: ")


if operator == "+":
    print(number1 + number2)
elif operator == "-":
    print(number1 - number2)
elif operator == "*":
    print(number1 * number2)
elif operator == "/":
    print(number1 / number2)
elif operator == "%":
    print(number1 % number2)
elif operator == "**":
    print(number1 ** number2)
elif operator == "//":
    print(number1 // number2)

else:
    print("Invalid operator")
```

```python
minAttendees = 20
ticketPrice = 10
fixedCost = 200
advertisingDollars = 0
maxProfit = 0
tempAd = 0
print("Ad$   \t  Profit")
for i in range(advertisingDollars, 225, 25):
  advertisingDollars = i
  additionalAttendees = (2 * round(advertisingDollars ** 0.5))
  totalAttendees = minAttendees + additionalAttendees
  profit = (totalAttendees * ticketPrice) - fixedCost - advertisingDollars
  print(f"{advertisingDollars}  \t  ${profit:,.2f}")
  if profit > maxProfit:
    maxProfit = profit
    tempAd = i
print(f"Highest profit is ${maxProfit:,.2f} when we spend ${tempAd} on Ads ")



```

```python
import random

play = input("Play game? (yes/no)").lower()

while play == 'yes':
  player = input("Choose Rock, Paper, or Scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player}\nThe Computer Chooses: {computer}")

  if player == computer:
    print("Tie")
  elif (
      (player ==  'rock' and computer == 'scissors') or
      (player == 'paper' and computer == 'rock') or
      (player == 'scissors' and computer == 'paper')
  ):
    print("You win")
  else:
    print("You lose")

  play = input("Play again? (yes/no)").lower()

```

```python
weights = [10.0,9.9,10.1,10.05]
total = 0
max = 0
for i in range(len(weights)):
  total += weights[i]
  if weights[i] > max:
    max = weights[i]
avg = total / len(weights)
print(f"Average = {avg}")
print(f"Max= {max}")


```
